#### Que.7 Convert list of nested dictionary into Pandas dataframe 

```python
import pandas as pd
l = [{"Student": [{"Exam": 90, "Grade": "a"},
                    {"Exam": 99, "Grade": "b"},
                    {"Exam": 97, "Grade": "c"}],
        "Name": "Paras Jain"},
        {"Student": [{"Exam": 89, "Grade": "a"},
                    {"Exam": 80, "Grade": "b"}],
        "Name": "Chunky Pandey"}]
print('list is: \n', l)
rows = []
for data in l:
    data_row = data['Student']
    time = data['Name']
    for row in data_row:
        row['Name']= time
        rows.append(row)
df = pd.DataFrame(rows)
print('\nDataframe is: \n')
df
```

```python
df = df.pivot_table(index ='Name', columns =['Grade'],
                        values =['Exam']).reset_index()
df.columns =['Name', 'Maths', 'Physics', 'Chemistry']
df
```

#### Que.13 Mapping external values to dataframe values in Pandas 

```python
import pandas as pd
data = {'First_name': ['Ram', 'Mohan', 'Tina', 'Jeetu', 'Meera'], 
        'Last_name': ['Kumar', 'Sharma', 'Ali', 'Gandhi', 'Kumari'], 
        'Age': [42, 52, 36, 21, 23], 
        'City': ['Mumbai', 'Noida', 'Pune', 'Delhi', 'Bihar']}
df = pd.DataFrame(data, columns = ['First_name', 'Last_name',
                                                      'Age', 'City'])
new_data = { "Ram":"B.Com",
             "Mohan":"IAS",
             "Tina":"LLB",
             "Jeetu":"B.Tech",
             "Meera":"MBBS" }
df["Qualification"] = df["First_name"].map(new_data)
df
```

```python
new_data = { "Ram":"Shyam",
             "Tina":"Riya",
             "Jeetu":"Jitender" }
df = df.replace({"First_name":new_data})
df
```

```python
new_data = { 0:"Shyam",
             2:"Riya",
             3:"Jitender" }
df["First_name"].update(pd.Series(new_data))
df
```
