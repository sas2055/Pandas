#### Que.1 How to iterate over rows in Pandas Dataframe 

```python
import pandas as pd
d = [{'name':'Sujit', 'age':27},{'name':'Sameer', 'age':18},
            {'name':'Sumit', 'age':22}]
 
df = pd.DataFrame(d)
df
```

#### Using iterrows

```python
for i, row in df.iterrows():
    print(row['name'], row['age'])
```

#### Que.2 Different ways to iterate over rows in Pandas Dataframe.

```python
import pandas as pd
data = {'Name': ['Ankit', 'Amit',
                 'Aishwarya', 'Priyanka'],
        'Age': [21, 19, 20, 18],
        'Stream': ['Diploma', 'Commerce',
                   'Arts', 'Science'],
        'Percentage': [88, 92, 95, 70]}
df = pd.DataFrame(data, columns=['Name', 'Age', 
                                 'Stream', 'Percentage'])
df
```

#### Using index

```python
for i in df.index:
    print(df['Name'][i], df['Stream'][i])
```

#### Using len

```python
for i in range(len(df)):
    print(df.loc[i, 'Name'], df.loc[i, 'Age'])
```

#### Using iterrows

```python
for i, row in df.iterrows():
    print(row['Name'], row['Percentage'])
```

#### Que.8 Select any row from a Dataframe using iloc[] and iat[] in Pandas 

#### using iloc[]

```python
import pandas as pd
df = pd.DataFrame({'Date':['10/2/2011', '11/2/2011', '12/2/2011', '13/2/11'],
                    'Event':['Music', 'Poetry', 'Theatre', 'Comedy'],
                    'Cost':[10000, 5000, 15000, 2000]})
ls =[]
for i in range((df.shape[0])):
    ls.append(list(df.iloc[i, :]))
print('list from rows: \n', ls[:3])
```

#### using iat[]

```python
import pandas as pd
df = pd.DataFrame({'Date':['10/2/2011', '11/2/2011', '12/2/2011', '13/2/11'],
                    'Event':['Music', 'Poetry', 'Theatre', 'Comedy'],
                    'Cost':[10000, 5000, 15000, 2000]})
ls = []
for i in range((df.shape[0])):
    row = []
    for j in range(df.shape[1]):
          row.append(df.iat[i, j])
    ls.append(row)
print('list from rows: \n', ls[:3])    
```

#### Que.10 Insert row at given position

```python
import pandas as pd
df = pd.DataFrame({'Date':['10/2/2011', '12/2/2011', '13/2/2011', '14/2/2011'],
                    'Event':['Music', 'Poetry', 'Theatre', 'Comedy'],
                    'Cost':[10000, 5000, 15000, 2000]})
df
```

```python
def Insert_row_(row_number, df, row_value):
    df1 = df[0:row_number]
    df2 = df[row_number:]
    df1.loc[row_number] = row_value
    df_result = pd.concat([df1, df2])
    df_result.index = [*range(df_result.shape[0])]
    return df_result
row_number = 2
row_value = ['11/2/2011', 'Wrestling', 12000]
if row_number > df.index.max()+1:
    print("Invalid row_number")
else:
    df = Insert_row_(2, df, row_value)
    print(df)
```
